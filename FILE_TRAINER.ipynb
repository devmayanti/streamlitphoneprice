# **DATA PREPARATION**

**Buat Kolom Infromasi Baru**

---



```python
# Membagi kolom 'Details' dan mengambil bagian pertama sebagai 'Storage'
data['Storage'] = data['Details'].str.split(' - ').str[0]

# Menghapus bagian yang telah diambil dari kolom 'Details'
data['Details'] = data['Details'].str.replace(r'^\S+ \S+ - ', '', regex=True)
data.head()
```

**Missing Values**

---



```python
# Convert 'current_price' from string to numeric
data['Selling price'] = data['current_price'].replace({'Rp': '', ',': ''}, regex=True).astype(float)

# Convert 'old_price' from string to numeric
data['Present price'] = data['old_price'].replace({'Rp': '', ',': ''}, regex=True).astype(float)

# Drop the original columns
data1 = data.drop(columns=['current_price', 'old_price'])
data1.head()
```

```python
# Ensure 'Storage' column is of string type before applying string methods
data1['Storage'] = data1['Storage'].astype(str)

# Replace 'GB' and 'TB' with empty string and convert to numeric
data1['Storage'] = data1['Storage'].str.replace('GB', '').str.replace('TB', '').astype(float)

# Handle 'TB' values (assuming 1 TB = 1024 GB)
# Changed data['Storage'] to data1['Storage'] for comparison
data1.loc[data1['Storage'] <= 10, 'Storage'] = data1.loc[data1['Storage'] <= 10, 'Storage'] * 1024
data1
```

```python
def data1_check(data1):
    missing = data1.isnull().sum()
    per_mis = 100*(missing/len(data1))
    data_type = data1.dtypes
    number_unique = data1.nunique()
    return pd.DataFrame({"Missing" : missing,
                        "Percent_Missing" : per_mis,
                        "data_type": data_type,
                        "number_unique" : number_unique})
data1_check(data1)
```

# **MODELLING**


**Splitting Data**

------

menggunakan stratified shuffle split untuk membagi data menjadi data train dan data test


```python
X = datatr.drop('Selling price', axis=1)
y = datatr['Selling price']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

**Set Cross Validation**

---



Cross validation yang digunakan pada pemodelan ini adalah repeated stratified k fold cross validation dengan 10 fold dan 8 repeat. pemilihan metode ini dikarenakan jumlah dataset yang sedikit dan untuk menghindari terjadinya overfitting

```python
# Import necessary libraries
from sklearn.model_selection import RepeatedKFold, cross_val_score

# Definisikan model
model = RandomForestRegressor(random_state=42, n_jobs=-1)

# Loop untuk berbagai nilai n_repeats
results = {}
for n_repeats in range(4, 9):
    cv = RepeatedKFold(n_splits=10, n_repeats=n_repeats, random_state=42)
    scores = cross_val_score(model, X, y, cv=cv, scoring='r2', n_jobs=-1)
    mean_score = np.mean(scores)
    results[f'n_repeats={n_repeats}'] = mean_score
    print(f'n_splits=10, n_repeats={n_repeats} -> R^2: {mean_score:.4f}')

# Menentukan kombinasi terbaik
best_repeats = max(results, key=results.get)
print(f'Kombinasi terbaik: {best_repeats} dengan R^2 tertinggi {results[best_repeats]:.4f}')
```

```python
# Define the cross-validation strategy
cv = RepeatedKFold(n_splits=10, n_repeats=4, random_state=42)
```

**Hyperparameter Tuning dengan GridSearchCV**

---



```python
# Hyperparameter Tuning with GridSearchCV
param_grid = {
    'n_estimators': [300, 400, 500],       # Lebih banyak variasi jumlah pohon
    'max_depth': [5, 20, 35],            # Kedalaman pohon yang lebih rendah dan lebih tinggi
    'min_samples_split': [4, 7, 15],            # Lebih banyak opsi split
    'min_samples_leaf': [1, 3, 5],             # Tambahkan lebih banyak variasi sampel minimum pada daun
}

grid_search = GridSearchCV(estimator=model, param_grid=param_grid,
                           scoring='neg_mean_absolute_error', n_jobs=-1, cv=cv, verbose=1)

# Fit GridSearchCV on the processed training data
grid_search.fit(X_train, y_train)

# Best parameters and model
print("Best parameters:", grid_search.best_params_)
best_model = grid_search.best_estimator_
best_model
```

```python
# Evaluasi model pada data pelatihan
y_train_pred = best_model.predict(X_train)
mae_train = mean_absolute_error(y_train, y_train_pred)
mse_train = mean_squared_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)

# Menghitung MAPE untuk data pelatihan
mape_train = np.mean(np.abs((y_train - y_train_pred) / y_train)) * 100

print(f'Training Set Evaluation:\n MAE: {mae_train:.2f}\n MSE: {mse_train:.2f}\n R^2: {r2_train:.2f}\n MAPE: {mape_train:.2f}%')
```

```python
# Evaluasi model pada data testing
y_test_pred = best_model.predict(X_test)
mae_test = mean_absolute_error(y_test, y_test_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
r2_test = r2_score(y_test, y_test_pred)
mape_test = np.mean(np.abs((y_test - y_test_pred) / y_test)) * 100

print(f'Test Set Evaluation:\n MAE: {mae_test:.2f}\n MSE: {mse_test:.2f}\n R^2: {r2_test:.2f}\n MAPE: {mape_test:.2f}%')
```
